```python
import pandas as pd
```

```python
dat_08 = pd.read_csv('cces_2008_common.tab', sep='\t')
```

```python
dat_08.head()
```

```python
import xml.etree.cElementTree as ET
tree = ET.ElementTree(file='cces_2008_common-ddi.xml')
root = tree.getroot()
namespace = '{http://www.icpsr.umich.edu/DDI}'
```

```python
data_descr = root.find(namespace+'dataDscr')
```

```python
list(root)
```

```python
list(data_descr)
```

```python
col_names = []
for col in dat_08.columns:
    print(col)
    value_dict = {}
    var_name = data_descr.find(namespace+'var[@name="'+col+'"]')
    lab_name = var_name.find(namespace+'labl')
    if var_name.find(namespace+'catgry') is not None:
        categories = var_name.findall(namespace+'catgry')
        for cat in categories:
            value_dict[cat.find(namespace+'catValu').text] = cat.find(namespace+'labl').text
        dat_08[col] = dat_08[col].astype('Int32').astype(str).map(value_dict)
    if lab_name.attrib['level']=='variable':
        dat_08.rename({col: lab_name.text}, axis=1, inplace=True)
        
    
```
