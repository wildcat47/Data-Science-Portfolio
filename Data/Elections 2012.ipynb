```python
import pandas as pd
```

```python
dat_12 = pd.read_csv('CCES12_Common_VV.tab', sep='\t')
```

```python
dat_12.head()
```

```python
import xml.etree.cElementTree as ET
tree = ET.ElementTree(file='CCES12_Common_VV-ddi.xml')
root = tree.getroot()
namespace = '{http://www.icpsr.umich.edu/DDI}'
```

```python
data_descr = root.find(namespace+'dataDscr')
```

```python
list(root)
```

```python
list(data_descr)
```

```python
col_names = []
for col in dat_12.columns:
    print(col)
    value_dict = {}
    var_name = data_descr.find(namespace+'var[@name="'+col+'"]')
    lab_name = var_name.find(namespace+'labl')
    if var_name.find(namespace+'catgry') is not None:
        categories = var_name.findall(namespace+'catgry')
        for i, cat in enumerate(categories):
            if (i==0) and ((cat.find(namespace+'labl').text=='Not Asked') or (cat.find(namespace+'labl').text=='Skipped')):
                break
            value_dict[cat.find(namespace+'catValu').text] = cat.find(namespace+'labl').text
        if len(value_dict.keys())>2:
            try:
                dat_12[col] = dat_12[col].astype('Int32').astype(str).map(value_dict)
            except:
                continue
    if lab_name.attrib['level']=='variable':
        dat_12.rename({col: lab_name.text}, axis=1, inplace=True)
        
    
```
